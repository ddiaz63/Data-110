```python
# Setup and Styles
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print(plt.style.available)
import seaborn as sns
sns.set_style("whitegrid")
plt.style.use('ggplot')
```

```python
#Basic Line Plot
x = np.linspace(-2,4,100)
y = x**2-2*x+1
plt.plot(x,y)
plt.title("Quad Function")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Multiple Lines
x1 = np.linspace(-2,2,100)
x2 = np.linspace(0.01,3,100)

f = x1**3
g = np.exp(x1)
h = np.log(x2)


plt.plot(x1, f, label='$f(x) = x^3$', color='pink')
plt.plot(x1, g, label='$g(x) = e^x$', color='purple')
plt.plot(x2, h, label='$h(x) = \log(x)$', color='blue')

# Add titles and labels
plt.title('Multiple Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Styling lines
x = np.linspace(-1,1,100)

f = x**2
g = x**3
h = x**4
i = x**5

plt.plot(x,f, label = 'x^2 graph', color = 'green', linestyle = '-')
plt.plot(x,g,label = 'x^3 graph', color = 'y', linestyle = '--')
plt.plot(x,h, label = 'x^4 graph', color = 'orange', linestyle = '-.')
plt.plot(x,i, label = 'x^5 graph', color = '#FF5744', linestyle = ':')

plt.title('Functions Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
#Basic Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.plot(x, y, 'o', label='Scatter with plt.plot()', color='blue')
plt.scatter(x,y)

plt.title('Basic Scatter Plot')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()
```

```python
#Customizing Scatter Plots
x = np.linspace(0, 4 * np.pi, 100)
y = x * np.sin(x)

plt.plot(x, y, marker='o', linestyle='-',
         markersize=8,
         linewidth=2,
         markerfacecolor='cyan',
         markeredgecolor='blue',
         markeredgewidth=1.5)

plt.title('Scatter Plot of $f(x) = x \sin(x)$')
plt.xlabel('x values')
plt.ylabel('f(x) values')
```

```python
# Combined Plot
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
f_x = x * np.cos(x)

scatter_x = np.arange(-2 * np.pi, 2 * np.pi + np.pi / 4, np.pi / 4)
scatter_y = scatter_x * np.cos(scatter_x)

plt.plot(x, f_x, label='$f(x) = x \cos(x)$', color='green', linestyle='-', linewidth=2)

plt.scatter(scatter_x, scatter_y, color='red', marker='o', s=100, label='Scatter points', edgecolor='black')

plt.title('Combined Plot of $f(x) = x \cos(x)$ with Scatter Points')
plt.xlabel('x values')
plt.ylabel('f(x) values')
plt.legend()
```

```python
#Custom Plot with Alternative Style

# Set the alternative style
plt.style.use('ggplot')

# Define the range for x, avoiding points where tan(x) is undefined
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x = x[np.abs(x) % (np.pi / 2) != 0]  # Remove points where tan(x) is undefined

# Define the functions
f_x = np.tan(x)        # f(x) = tan(x)
i_x = np.arctan(x)     # i(x) = arctan(x)

# Create the custom plot
plt.figure(figsize=(12, 8))

# Plot f(x) with a solid line
plt.plot(x, f_x, label='$f(x) = \tan(x)$', color='blue', linestyle='-', linewidth=2)

# Plot i(x) with a dashed line
plt.plot(x, i_x, label='$i(x) = \arctan(x)$', color='orange', linestyle='--', linewidth=2)

# Scatter points for i(x) at specific intervals
scatter_x = np.arange(-5, 6, 1)  # Scatter points at integer values
scatter_y = np.arctan(scatter_x)  # i(x) values at scatter points
plt.scatter(scatter_x, scatter_y, color='red', marker='o', s=100, label='Scatter points', edgecolor='black')

# Customize the plot
plt.title('Custom Plot of $f(x) = \\tan(x)$ and $i(x) = \\arctan(x)$')
plt.xlabel('x values')
plt.ylabel('Function values')
plt.xlim(-7, 7)  # Set x limits
plt.ylim(-10, 10)  # Set y limits for tan(x)
plt.legend()
```

```python

# Generate data
x = np.linspace(-2, 4, 100)
y = x**2 - 2*x + 1

# ggplot style
plt.style.use('ggplot')
plt.plot(x, y)
plt.title("Quadratic Function (Default Style)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


# Seaborn-whitegrid style
plt.style.use('seaborn-whitegrid')
plt.plot(x, y)
plt.title("Quadratic Function (Seaborn-Whitegrid Style)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

# Default
plt.style.use('default')
plt.plot(x, y)
plt.title("Quadratic Function (GGPlot Style)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

The default style is the most appropriate choice for straightforward data visualizations where clarity and simplicity are paramount.
