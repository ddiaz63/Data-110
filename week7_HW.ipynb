```python
# Imported libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load iris dataset
iris = sns.load_dataset('iris')
```

```python
# 1.1: Create a pair plot with color-coding by species

sns.pairplot(iris, hue='species')
plt.suptitle("Pair Plot of Iris Dataset Features by Species", y=1.02)
plt.show()
```

The pair plot shows the relationships between different features, and we can observe clusters that differentiate species, particularly with petal length and petal width.

```python
# 1.2: Bar plot for average sepal length by species

sns.barplot(x='species', y='sepal_length', data=iris, ci=None)
plt.title("Average Sepal Length for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length")
plt.show()
```

We can see distinct differences in the average sepal lengths between species, particularly noting that setosa has the smallest average sepal length.

```python
# 1.3: Box plot for petal width distribution by species

sns.boxplot(x='species', y='petal_width', data=iris)
plt.title("Distribution of Petal Width for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Petal Width")
plt.show()
```

The box plot shows the spread and median of petal widths per species, where setosa has the smallest petal width, and virginica has the largest.

```python
# 1.4: Grouped bar plot for average sepal and petal lengths by species

species_means = iris.groupby('species')[['sepal_length', 'petal_length']].mean().reset_index()
species_means.plot(x='species', kind='bar')
plt.title("Average Sepal and Petal Lengths by Species")
plt.xlabel("Species")
plt.ylabel("Length (cm)")
plt.legend(["Sepal Length", "Petal Length"])
plt.show()
```

The grouped bar plot highlights that petal length varies more across species than sepal length, which could indicate a characteristic feature for classification.

```python
# 1.5: Violin plot for sepal length distribution by species

sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title("Distribution of Sepal Length for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.show()
```

The violin plot displays the distribution shapes for sepal length, showing that versicolor has a wider range and setosa has a smaller, more concentrated range.
