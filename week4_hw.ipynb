```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")  # Sets the style for the plots
```

```python
#Create a bar chart for crime rates in 2023.
#Customize the bar colors and add gridlines.
#Label the axes and add a title.

data = { 'Offense': ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
    'Offenses per 100,000 people (2023)': [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5],
    'Change in Offenses per 100,000 People (2022-2023)': [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]}

df = pd.DataFrame(data)
colors = ['#33FFF0', '#33FF57', '#FF33A1', '#3357FF', '#FFC300', '#FF33A1', '#FF5733']

plt.figure(figsize=(10, 6))
bars = plt.bar(df['Offense'], df['Offenses per 100,000 people (2023)'], color=colors)

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xlabel('Crime Type')
plt.ylabel('Offense Rate (per 100,000)')
plt.title('Crime Rates per 100,000 People in 2023')


plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
#Create a horizontal bar chart showing the change in offense rates.
#Assign a unique color to each crime type.
#Label the axes and add a title.

plt.figure(figsize=(10, 6))
bars = plt.barh(df['Offense'], df['Change in Offenses per 100,000 People (2022-2023)'], color=colors)

# Add gridlines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Label the axes
plt.xlabel('Percentage Change in Offenses (per 100,000)')
plt.title('Percentage Change in Crime Rates from 2022 to 2023')

# Adding data labels on bars
for index, value in enumerate(df['Change in Offenses per 100,000 People (2022-2023)']):
    plt.text(value, index, f'{value}%', va='center')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
bars = plt.barh(df['Offense'], df['Change in Offenses per 100,000 People (2022-2023)'], color=['blue' if x > 0 else 'pink' for x in df['Change in Offenses per 100,000 People (2022-2023)']])

# Add gridlines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Label the axes
plt.xlabel('Percentage Change in Offenses (per 100,000)')
plt.title('Percentage Change in Crime Rates from 2022 to 2023')

# Adding data labels on bars
for index, value in enumerate(df['Change in Offenses per 100,000 People (2022-2023)']):
    plt.text(value, index, f'{value}%', va='center')

plt.tight_layout()
plt.show()
```

**Why does this bar graph make more sense than Task 1's bar graph?**

A horizontal bar chart showing percentage changes clearly shows how crime rates are moving—whether they are going up or down, which is important for making good policy decisions. Unlike absolute numbers that can hide important details, percentage changes help compare different types of crime more easily. This way, we can spot trends that need attention, like increasing thefts or decreasing homicide rates, which can guide prevention efforts.
